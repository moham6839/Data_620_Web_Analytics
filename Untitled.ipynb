```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a Krackhardt kite graph
G = nx.krackhardt_kite_graph()

# Draw the graph
plt.figure(figsize=(10, 6))  # Ensure this is not overridden in your environment
nx.draw_networkx(
    G,
    with_labels=True,
    node_color="skyblue",
    node_size=1000,
    font_size=15,
    font_color="black",
    font_weight="bold",
    edge_color="gray",
)
plt.title("Krackhardt Kite Graph", fontsize=16)
plt.show()
```

```python
net.add_node(1, label="Node 1") # node id = 1 and label = Node 1
net.add_node(2) # node id and label = 2
```
